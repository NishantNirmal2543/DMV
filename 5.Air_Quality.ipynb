```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("City_Air_Quality.csv")

# Convert the 'date' column to a datetime format
df['date'] = pd.to_datetime(df['date'])

# 1. Identify relevant variables
date = df['date']
so2 = df['so2']
no2 = df['no2']
pm25 = df['pm2_5']
pm10 = df['spm']  # Using 'spm' for PM10 as no specific column is given
aqi = df['rspm']  # You can choose another column for AQI values if available

# 2. Create a line plot for the overall AQI trend over time
plt.figure(figsize=(12, 6))
plt.plot(date, aqi, label='AQI', color='blue')
plt.title('AQI Trend Over Time')
plt.xlabel('Date')
plt.ylabel('AQI Value')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()

# 3. Plot individual pollutant levels (SO2, NO2, PM2.5, PM10) on separate line plots
plt.figure(figsize=(12, 6))
plt.plot(date, so2, label='SO2', color='red')
plt.title('SO2 Trend Over Time')
plt.xlabel('Date')
plt.ylabel('SO2 Level')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(date, no2, label='NO2', color='green')
plt.title('NO2 Trend Over Time')
plt.xlabel('Date')
plt.ylabel('NO2 Level')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(date, pm25, label='PM2.5', color='purple')
plt.title('PM2.5 Trend Over Time')
plt.xlabel('Date')
plt.ylabel('PM2.5 Level')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(date, pm10, label='PM10', color='orange')
plt.title('PM10 Trend Over Time')
plt.xlabel('Date')
plt.ylabel('PM10 Level')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```
